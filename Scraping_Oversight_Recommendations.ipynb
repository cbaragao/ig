```python
import requests, re, math
import pandas as pd
from bs4 import BeautifulSoup

page = requests.get("https://www.oversight.gov/recommendations", verify=False)
soup = BeautifulSoup(page.content, 'html.parser')
rec_text = soup.findAll("div", {"class":"oversight-table-summary"})
rec_list = re.split("\s", rec_text[0].get_text()) 
number_recs = float(rec_list[5])
number_pages = math.ceil(number_recs/20)

recs = []
recoms = []
strip_recoms = []
final_recoms = []

def scrape_recommendations(webpage, page_number):
    next_page = webpage + str(page_number)
    response = requests.get(str(next_page))
    soup = BeautifulSoup(response.content, 'html.parser')
    
    recommendations = soup.find_all('span', class_="field-content")

    for el in recommendations:
        recs.append(el.get_text())
       
    if page_number < number_pages:
    #if page_number < 5:
        page_number = page_number + 1
        scrape_recommendations(webpage, page_number)

    return recs

def get_even_list_items(strip_list):
    for item, val in enumerate(strip_list):
        if val.isnumeric() == False and len(val) > 10 and not ('\nOpenRecommendations') in val:
            strip_recoms.append(val)
    return strip_recoms

recoms = scrape_recommendations('https://www.oversight.gov/recommendations?page=', 0)
final_recoms = get_even_list_items(recoms)

df = pd.DataFrame(final_recoms)

df.head(50)
```
